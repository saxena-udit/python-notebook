```python
my_port = load_portfolio_file('my_portfolio.txt')
```

```python
market_data = load_price_history(my_port['SYMBOL'],days=30,time_interval='1D')
```

```python
market_data_df = pd.DataFrame({})
for data in market_data:
    ticker_data = data.data
    ticker_data['PREV_c'] = ticker_data['c'].shift(1)
    ticker_data['CHANGE_c'] = ticker_data['c'] - ticker_data['PREV_c']
    #ticker_data.drop([0,1],inplace=True)
    market_data_df = market_data_df.append(data.data)
    
my_port_data = market_data_df.set_index('SYMBOL').join(my_port.set_index('SYMBOL'))
my_port_data['MKT_VALUE'] = my_port_data['QTY'] * my_port_data['c']
my_port_data['CHANGE'] = (my_port_data['MKT_VALUE'] - my_port_data['INVESTMENT'])*100 / my_port_data['INVESTMENT']
```

```python
my_port_data = my_port_data.reset_index()
```

```python
ticker_filter(my_port_data.sort_values(['SYMBOL','time']),'DCBBANK')
```

```python
fig = go.Figure()
for ticker in my_port_data['SYMBOL'].unique():
    ticker_data = ticker_filter(my_port_data,ticker)
    fig.add_trace(go.Scatter(x=ticker_data['time'], y=ticker_data['CHANGE'],
                        mode='lines+markers',
                        name=ticker))    
fig.show()
```

```python
ca_info = load_ca_info()
```

```python
div_info = ca_info
div_info['Ex-Date'] = pd.to_datetime(div_info['Ex-Date'], dayfirst=True)
div_info = div_info[div_info['Purpose'].str.contains('Dividend ')].set_index('SYMBOL').reset_index()

div_info["dividend"] = div_info["Purpose"].str.extract(r'.*R[s|e][\s.-]* ([0-9.]*)')
#div_info["dividend"] = div_info["dividend"].str.strip()
div_info["dividend"] = pd.to_numeric(div_info["dividend"], errors='coerce')
#div_info["dividend"].fillna(div_info["dividend_2"],inplace=True)
div_info[div_info["dividend"].isnull()]
```
