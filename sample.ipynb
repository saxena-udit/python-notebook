```python

file_types=["INDEX","CM","WDM","FO","SME","SLBS","COM","CD","NBF","CBM"]
file_days=['PreviousDay','FutureDay','CurrentDay']
report_urls = []
base_url = 'https://www.nseindia.com/api/daily-reports?key='
for file_type in file_types:
    data = requests.get(base_url+file_type,headers=chrome_headers,cookies=get_cookies()).json()
    for file_day in file_days:
        for file in data[file_day]:
            report_urls.append(MyURL(file['displayName'],file['filePath']+file['fileActlName'],'csv'))
        
```

```python
print('total url found',len(report_urls))
```

```python
report_urls.sort(key=lambda x: x.url)
for myurl in filter(lambda item: '.csv' in item.url , report_urls):
    try:
        loadURL(myurl,True)
    except Exception:
        print('Error loading URL ', myurl)
```

```python
empty_data_url = [a for a in report_urls if a.data is None]
url_with_data = [a for a in report_urls if a.data is not None]
```
