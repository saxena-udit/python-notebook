```python
%%capture
market_data = load_price_history(my_port['SYMBOL'],days=50,time_interval='1D')
```

```python
market_data_df = pd.DataFrame({})
for data in market_data:
    ticker_data = data.data
    ticker_data['PREV_c'] = ticker_data['c'].shift(1)
    ticker_data['CHANGE_c'] = ticker_data['c'] - ticker_data['PREV_c']
    #ticker_data.drop([0,1],inplace=True)
    market_data_df = market_data_df.append(data.data)
    
my_port_data = market_data_df.set_index('SYMBOL').join(my_port.set_index('SYMBOL'))
my_port_data = my_port_data[my_port_data['time'] > my_port_data['DATE']]
my_port_data['MKT_VALUE'] = my_port_data['QTY'] * my_port_data['c']
my_port_data['P/L'] = (my_port_data['MKT_VALUE'] - my_port_data['INVESTMENT'])
my_port_data['CHANGE'] = my_port_data['P/L'] * 100 / my_port_data['INVESTMENT']
```

```python
my_port_data = my_port_data.reset_index()
```

```python
dcb_data = ticker_filter(my_port_data.sort_values(['SYMBOL','time']),'NHPC')
dcb_data
```

```python
fig = go.Figure()
for ticker in my_port_data['SYMBOL'].unique():
    ticker_data = ticker_filter(my_port_data,ticker)
    ticker_data = ticker_data.groupby(['time','SYMBOL']).agg({'P/L':'sum'}).reset_index()
    fig.add_trace(go.Scatter(x=ticker_data['time'], y=ticker_data['P/L'],
                        mode='lines+markers',
                        name=ticker))

agg_data = my_port_data.groupby('time').agg({'P/L':'sum'}).reset_index()
fig.add_trace(go.Bar(x=agg_data['time'], y=agg_data['P/L'],
                        marker_color='indianred',
                        name='Total'))
fig.show()
```
